```python
import pandas as pd

# Create a sample dataset
data = {
    "rainfall": [800, 650, 900, 700, 850, 620, 780],
    "fertilizer": [120, 100, 150, 110, 140, 90, 130],
    "temperature": [25, 27, 24, 26, 25, 28, 26],
    "yield": [3.2, 2.8, 3.6, 3.0, 3.4, 2.6, 3.1]
}

df = pd.DataFrame(data)

# Add seed rate (kg/ha) and plant population (plants/ha)
df["seed_rate"] = [20, 18, 22, 19, 21, 17, 20]
df["plant_population"] = [180000, 165000, 195000, 170000, 185000, 160000, 175000]

print(df)
```

```python
import pandas as pd

data = {
    "Yield_kg": [3200, 3000, 3500, 2800, 3600, 2900, 3300, 3100, 3450, 2950,2030,2531,2235],
    "Soil_Moisture": [30, 28, 35, 25, 36, 27, 31, 29, 34, 26,32,25,24]
}
```

```python
import numpy as np

plots = np.array([30, 28, 35, 25, 36, 27, 31, 29, 34, 26, 32, 25, 24])
center = 30

plots - center
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
df["Zone"] = kmeans.fit_predict(df)
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Yield_kg"], df["Soil_Moisture"], c=df["Zone"])
plt.xlabel("Yield (kg)")
plt.ylabel("Soil Moisture")
plt.title("Farm Field Clustering")
plt.show()
```
