```python
# Equipment Inventory using Stack (LIFO)

inventory = []  # empty stack

# Add equipment (push)
def add_equipment(item):
    inventory.append(item)
    print(item, "added to inventory")

# Use equipment (pop)
def use_equipment():
    if len(inventory) == 0:
        print("No equipment available")
    else:
        item = inventory.pop()
        print(item, "used from inventory")

# Display inventory
def show_inventory():
    print("Current Inventory:", inventory)

# Example usage
add_equipment("Helmet")
add_equipment("Gloves")
add_equipment("Boots")
add_equipment("saw")
show_inventory()

use_equipment()
use_equipment()

show_inventory()
```
