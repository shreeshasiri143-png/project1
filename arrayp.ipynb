```python
import numpy as np

# 1. INPUTS
# Define the range for 5 levels of Nitrogen (rows) and Phosphorus (cols)
n_doses = np.arange(1, 6)  # Creates array: [1, 2, 3, 4, 5]
p_doses = np.arange(1, 6)  # Creates array: [1, 2, 3, 4, 5]

# 2. DATA STRUCTURE
# We will create a 5x5 NumPy ndarray to store the results.

# 3. PROCESSING
# Use NumPy broadcasting to multiply a column vector by a row vector
# 'None' creates a new axis to transform n_doses into a vertical vector
dosage_table = n_doses[:, None] * p_doses

# 4. OUTPUT
print("Fertilizer Dosage Table (N x P) - NumPy 2026")
print("-" * 45)
# Print Phosphorus header
print("N\\P |", "  ".join([f"{p:2}" for p in p_doses]))
print("-" * 25)

# Print each row with Nitrogen level labels
for i, row in enumerate(dosage_table, start=1):
    formatted_row = "  ".join([f"{val:2}" for val in row])
    print(f" {i}  | {formatted_row}")
```
