
# 🌱 Python Classes for Agriculture Students  
### Learning Object-Oriented Programming with Farming Examples



## 🤔 Why Learn Classes?

A **class** is like a **blueprint**.

🚜 Tractor blueprint → Actual tractor  
🌾 Crop blueprint → Actual crop record


```python

# Creating an empty class
class Crop:
    pass
```


## 🌾 Adding Crop Details (Attributes)


```python

class Crop:
    def __init__(self, name, season, expected_yield):
        self.name = name
        self.season = season
        self.expected_yield = expected_yield
```


## 🌱 Creating Crop Objects


```python

rice = Crop("Rice", "Kharif", 5.2)
wheat = Crop("Wheat", "Rabi", 4.1)

print(rice.name, rice.season, rice.expected_yield)
print(wheat.name, wheat.season, wheat.expected_yield)
```


## 🧑‍🌾 Adding Behavior (Methods)


```python

class Crop:
    def __init__(self, name, season, expected_yield):
        self.name = name
        self.season = season
        self.expected_yield = expected_yield

    def total_production(self, area):
        return self.expected_yield * area
```


## 📐 Using the Method


```python

area = 10
rice = Crop("Rice", "Kharif", 5.2)
print("Total Rice Production:", rice.total_production(area), "tons")
```


## 🚜 Mapping Farming to Python

| Farming | Python |
|-------|--------|
Crop Type | Class |
Field Record | Object |
Yield | Attribute |
Production Formula | Method |
