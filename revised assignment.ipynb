```python
file=open("assignment.txt","a")
file.write("\n using pop function")

inventory = [] # Initialize the inventory list

def add_equipment(item):
    inventory.append(item)
    print(item, "added to inventory")

# Use equipment (pop)
def use_equipment():
    if len(inventory) == 0:
        print("No equipment available")
    else:
        item = inventory.pop()
        print(item, "used from inventory")

# Display inventory
def show_inventory():
    print("Current Inventory:", inventory)

# Example usage
add_equipment("Helmet")
add_equipment("Gloves")
add_equipment("Boots")
add_equipment("saw")
show_inventory()

use_equipment()
use_equipment()

show_inventory()

file.close()
```

```python
import array

# 'i' represents signed integers (Equipment IDs)
inventory_array = array.array('i', [])

def push_equipment(item_id):
    """Adds a numeric ID to the stack."""
    inventory_array.append(item_id)
    print(f"Pushed ID: {item_id}")

def pop_equipment():
    """Removes the last numeric ID added."""
    if len(inventory_array) == 0:
        return "Error: Stack is empty"

    # .pop() works on array objects just like lists
    removed = inventory_array.pop()
    return f"Popped ID: {removed}"

# --- Testing the Array Stack ---
push_equipment(101)
push_equipment(102)
push_equipment(103)

print("Current Stack:", inventory_array)

print(pop_equipment()) # Will remove 103
print("Remaining:", inventory_array)
```

```python
# Equipment Inventory Stack using List (LIFO)

stack = []   # stack to store equipment names

# -------- INPUT --------
n = int(input("Enter number of equipment items to add: "))

for i in range(n):
    item = input(f"Enter equipment {i+1}: ")
    stack.append(item)

# -------- PROCESSING --------
used_item = None
if stack:
    used_item = stack.pop()   # LIFO operation

# -------- OUTPUT TO TEXT FILE --------
with open("assignment.txt", "w") as file:
    file.write("Equipment Inventory System (LIFO)\n")
    file.write("---------------------------------\n")
    file.write("Equipment added to inventory:\n")

    for item in stack:
        file.write(item + "\n")

    file.write("\n")
    if used_item:
        file.write(f"Equipment used (last added): {used_item}\n")
    else:
        file.write("No equipment available to use.\n")

print("Inventory details saved to assignment.txt")
```

```python
def compute_emi(P,R,N):
  r=R/(12*100)
  n=N*12
  emi=(P*r*(1+r)**n)/((1+r)**n-1)
  return emi
P=float(input("principal:"))
S=float(input("annual interest rate(%):"))
N=int(input("tenure(years):"))
emi_value=compute_emi(P,S,N)
print("monthly emi:",emi_value)
```
