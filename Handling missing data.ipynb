```python
import pandas as pd
import numpy as np
farm_data = {
"Crop": ["Wheat", "Rice", "Maize", "Millet", "Pulses"],
"Yield_kg": [3200, np.nan, 3500, 3000, np.nan],
"Water_liters": [4200, 5500, np.nan, 3000, 2500],
"Cost_INR": [18000, 20000, 17000, np.nan, 12000]
}
df = pd.DataFrame(farm_data)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df["Yield_kg"].fillna(df["Yield_kg"].mean(), inplace=True)
df
```

```python
df.fillna(method="ffill", inplace=True)
df
```

```python
df.fillna(method="bfill", inplace=True)
df
```

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Convert to DataFrame
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = iris.target_names[y]

print(df.head())
print(f"\nShape: {df.shape}")
```

```python
df.isnull()
```

```python
df.isnull().sum()
```
