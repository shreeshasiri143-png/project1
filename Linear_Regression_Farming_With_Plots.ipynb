# 🌾 Linear Regression for Agriculture (With Plots)
Google Colab Compatible

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Simple Linear Regression: Rainfall vs Yield

```python

data = {
    "Rainfall_mm": [180, 190, 200, 210, 220],
    "Yield_kg": [3000, 3100, 3200, 3400, 3500]
}
df = pd.DataFrame(data)
df
```

```python

X = df[['Rainfall_mm']]
y = df['Yield_kg']
model = LinearRegression()
model.fit(X, y)
```

```python

plt.figure()
plt.scatter(df['Rainfall_mm'], df['Yield_kg'])
plt.plot(df['Rainfall_mm'], model.predict(X))
plt.xlabel("Rainfall (mm)")
plt.ylabel("Yield (kg)")
plt.title("Rainfall vs Yield")
plt.show()
```

## Multiple Linear Regression: Rainfall + Fertilizer

```python

data_multi = {
    "Rainfall_mm": [180, 190, 200, 210, 220],
    "Fertilizer_kg": [40, 45, 50, 55, 60],
    "Yield_kg": [3000, 3150, 3300, 3450, 3600]
}
df_multi = pd.DataFrame(data_multi)
df_multi
```

```python

X = df_multi[['Rainfall_mm', 'Fertilizer_kg']]
y = df_multi['Yield_kg']
model_multi = LinearRegression()
model_multi.fit(X, y)
model_multi.coef_
```
