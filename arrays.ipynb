```python
class EquipmentInventory:
    def __init__(self, capacity):
        # Initialize the inventory with a fixed capacity using a Python list
        # This simulates a fixed-size array as implied by 'CREATE array STACK[capacity]'
        self.capacity = capacity
        self.stack = [None] * capacity
        self.top = -1 # 'top' index, -1 indicates empty

    def add_equipment(self, item):
        if self.top == self.capacity - 1:
            print("Inventory Full")
        else:
            self.top += 1
            self.stack[self.top] = item
            print(f"'{item}' added to inventory.")

    def use_equipment(self):
        if self.top == -1:
            print("Inventory Empty")
            return None # Indicate no item was used
        else:
            item = self.stack[self.top]
            self.stack[self.top] = None # Optionally clear the slot, not strictly necessary for LIFO logic
            self.top -= 1
            print(f"'{item}' used from inventory.")
            return item

    def display_inventory(self):
        if self.top == -1:
            print("Inventory Empty")
        else:
            print("Current Inventory (top to bottom):")
            # Iterate from the 'top' index down to 0
            for i in range(self.top, -1, -1):
                print(f"- {self.stack[i]}")

# --- Example Usage ---
# Let's set a capacity for our inventory
# (The original pseudocode used 'n', so we define it here for the example)
n = 3 # Example capacity
inventory = EquipmentInventory(n)

print("--- Adding Equipment ---")
inventory.add_equipment("Hammer")
inventory.add_equipment("Screwdriver")
inventory.add_equipment("Wrench")
inventory.add_equipment("Drill") # This should show 'Inventory Full'

print("\n--- Displaying Inventory ---")
inventory.display_inventory()

print("\n--- Using Equipment ---")
used_item = inventory.use_equipment()
used_item = inventory.use_equipment()

print("\n--- Displaying Inventory After Use ---")
inventory.display_inventory()

print("\n--- Adding More Equipment ---")
inventory.add_equipment("Pliers")

print("\n--- Final Display ---")
inventory.display_inventory()

print("\n--- Emptying Inventory ---")
inventory.use_equipment()
inventory.use_equipment()
inventory.use_equipment() # This should show 'Inventory Empty'
```
