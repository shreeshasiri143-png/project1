# 📊 Advanced Data Visualization for Agricultural Analysis
From Raw Farm Records to Actionable Insights

```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## 🌾 Creating a Sample Agricultural Dataset

```

np.random.seed(42)

data = pd.DataFrame({
    "Crop_Yield_tons": np.random.normal(4.5, 0.8, 120),
    "Water_Usage_mm": np.random.normal(520, 90, 120),
    "Cultivation_Cost_INR": np.random.normal(38000, 7000, 120),
    "Fertilizer_kg": np.random.normal(210, 35, 120)
})

data.head()
```

## 📊 Histogram – Crop Yield Distribution

```

plt.figure(figsize=(8,5))
plt.hist(data["Crop_Yield_tons"], bins=15, edgecolor="black")

plt.xlabel("Crop Yield (tons per hectare)")
plt.ylabel("Number of Farms")
plt.title("Distribution of Crop Yield Across Farms")

plt.show()
```

## 📦 Box Plot – Yield Risk & Outliers

```

plt.figure(figsize=(7,5))
sns.boxplot(y=data["Crop_Yield_tons"])

plt.ylabel("Crop Yield (tons per hectare)")
plt.title("Box Plot of Crop Yield – Risk & Outliers")

plt.show()
```

## 🔥 Heatmap – Resource Correlations

```

correlation = data.corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap="YlGnBu", fmt=".2f")

plt.title("Heatmap: Relationship Between Agricultural Variables")
plt.show()
```
