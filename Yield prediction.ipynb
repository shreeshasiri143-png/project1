```python
# Step 1: Import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Step 2: Create dataset
data = {
    "Rainfall":[800,600,400,700,900,300],
    "Temperature":[25,30,35,28,24,36],
    "Yield":["High","Medium","Low","Medium","High","Low"]
}

df = pd.DataFrame(data)

# Step 3: Separate features and target
X = df[["Rainfall","Temperature"]]
y = df["Yield"]

# Step 4: Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 5: Train Decision Tree model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Step 6: Make predictions
y_pred = model.predict(X_test)

# Step 7: Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)

# Step 8: Predict new data
new_data_input_1 = [[850, 26]]   # Rainfall, Temperature
new_data_df_1 = pd.DataFrame(new_data_input_1, columns=X_train.columns)
prediction = model.predict(new_data_df_1)

print("Predicted Yield:", prediction)
# Enter new data
rainfall = float(input("Enter Rainfall (mm): "))
temperature = float(input("Enter Temperature (°C): "))

# Make prediction
new_data_input_2 = [[rainfall, temperature]]
new_data_df_2 = pd.DataFrame(new_data_input_2, columns=X_train.columns)
prediction = model.predict(new_data_df_2)

# Print result
print("Predicted Yield:", prediction[0])
```
