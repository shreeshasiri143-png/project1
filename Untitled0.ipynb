```python
def find_min_marks(marks_list):
    """Finds the minimum mark in a given list of marks."""
    if not marks_list:
        return None # Handle empty list case
    return min(marks_list)

# Example usage of the function:
my_marks = [56, 78, 90, 45, 88]
minimum_value = find_min_marks(my_marks)

if minimum_value is not None:
    print(f"The minimum mark in the list {my_marks} is: {minimum_value}")
else:
    print("The list is empty.")
```
